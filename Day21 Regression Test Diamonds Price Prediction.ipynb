```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(df)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

# 1 Linear Regreesiom

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
```

```python
# Sum of Square Error
sse=sum((y_pred1-y_test)**2)
print('Sum of Squared Error',sse)

# mean square error
mse=sum((y_pred1-y_test)**2)/len(y_test)
print('mean of Squared Error',mse)

# root mean square error

rmse=np.sqrt(sum((y_pred1-y_test)**2)/len(y_test))
print(' Root mean of Squared Error',rmse)
```

#  2 Ridge Regression L2

```python
from sklearn.linear_model import Ridge
rid=Ridge()
rid.fit(x_train,y_train)
y_pred2=rid.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
# root mean square error

rmse = np.sqrt(mean_squared_error(y_test, y_pred2))
print("Root Mean Squared Error:", rmse)
```

#   3 Lasso Regression L1

```python
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(x_train,y_train)
y_pred3=ls.predict(x_test)
```

```python
# root mean square error

rmse = np.sqrt(mean_squared_error(y_test, y_pred3))
print("Root Mean Squared Error:", rmse)
```

# 4 Elastic Net Regression

```python
from sklearn.linear_model import ElasticNet

elk=ElasticNet()  
elk.fit(x_train, y_train)
y_pred4 = elk.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred4))
print("Root Mean Squared Error:", rmse)
```

# 5 Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
dt = DecisionTreeRegressor()
param_dict = {
    'criterion': ["squared_error", "friedman_mse", "absolute_error"],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 5, 10, 20]}

grid = GridSearchCV(estimator=dt, param_grid=param_dict, cv=5)
grid.fit(x_train, y_train)
print("Best Parameters:", grid.best_params_)

# Use best estimator for prediction
best_dt = grid.best_estimator_
y_pred5 = best_dt.predict(x_test)
```
